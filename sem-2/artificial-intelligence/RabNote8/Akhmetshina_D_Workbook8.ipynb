```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

iris = load_iris()

X1 = iris.data[:, [0, 1]]
X2 = iris.data[:, [2, 3]]

Clusters_count = 2
kmeans = KMeans(n_clusters = Clusters_count, random_state=0)
Y1 = kmeans.fit_predict(X1)
Y2 = kmeans.fit_predict(X2)
plt.scatter(X1[:, 0], X1[:, 1], c = Y1, cmap = 'summer' )
plt.scatter(X2[:, 0], X2[:, 1], c = Y2, cmap = 'rainbow' )
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc



iris = load_iris()

X = iris.data[:, [0, 1]]

plt.figure(figsize=(28, 12), dpi=180)
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(X, method = 'ward'))

plt.show()

cluster = AgglomerativeClustering(n_clusters = 2, affinity='euclidean', linkage = 'ward')
Y = cluster.fit_predict(X)
plt.figure(figsize= (10, 7))

plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = 'rainbow' )
```
